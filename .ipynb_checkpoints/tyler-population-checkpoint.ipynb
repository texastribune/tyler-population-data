```python
# set census api key 
c = Census("8149a6071641f66c9a2eaa2a679734a3d2a148e1")
# we want all fips codes for texas
state_fips = states.TX.fips
# want data for the past 20 years 
years = list(range(2004, 2025)) 
```

```python
# looking at metropolitan statistical areas
msa_data = [] 

# loop through to get population data for all MSAs in texas using ACS 1-year estimates.
# note that data is not available for 2020, and the earliest data available for ACS 1-year is 2005 (not 2004)
for year in years:
    try:
        # Get MSA data - no state filter is needed as MSAs are their own geography
        data = c.acs1.get(
            ('NAME', 
             'B01003_001E', # total population
             'B01002_001E', # median age
             'B09001_001E', # population under 18
             'B07003_004E', # Lived in the same house 1 year ago
             'B07003_007E', # Moved within same county
             'B07003_010E', # Moved from a different county in the same state
             'B07003_013E', # Moved from a different state
             'B07003_016E', # Moved from abroad 
            ), 
            geo={'for': 'metropolitan statistical area/micropolitan statistical area:*'},
            year=year
        )
        
        # Add year to each record
        for record in data:
            record['year'] = year
            msa_data.append(record)
        
        print(f"Retrieved data for {year}: {len(data)} MSAs")
    
    except Exception as e:
        print(f"Error for year {year}: {e}")
# make it a dataframe
df_msa = pd.DataFrame(msa_data)

# filter for only MSAs in Texas
msa_tx = df_msa[df_msa['NAME'].str.contains('TX', case=False, na=False)]

# rename cols
msa_tx = msa_tx.rename(columns={
    'NAME':'name',
    'B01003_001E': 'population',
    'B01002_001E': 'median_age',
    'B09001_001E': 'population_under_18',
    'B07003_004E': 'lived_in_same_house',
    'B07003_007E': 'moved_within_county',
    'B07003_010E': 'moved_from_diff_county_same_state',
    'B07003_013E': 'moved_from_diff_state',
    'B07003_016E': 'moved_from_abroad',
    'metropolitan statistical area/micropolitan statistical area':'msa_code'
}
) 
```

```python
msa_tx['msa_code'].nunique()
```

```python
# some msa_codes have multiple names
msa_tx.groupby('msa_code')['name'].unique()
# fix that by aggregating instances where there are multiple rows for the same metro area for the same year
# pick just the most common name 
msa_tx['name'] = msa_tx.groupby('msa_code')['name'].transform(lambda x: x.mode()[0])

msa_tx_agg = msa_tx.groupby(['msa_code', 'name', 'year']).agg({
    'population': 'sum',
    'median_age': 'median',        
    'population_under_18': 'sum'
}).reset_index()
```

```python
# pivot df so each column is a year
msa_tx_pop = msa_tx_agg.pivot(
    index=['msa_code','name'], 
    columns='year', 
    values='population'
).reset_index()

# pivot df for median age 
msa_tx_age = msa_tx_agg.pivot(
    index=['msa_code','name'], 
    columns='year', 
    values='median_age'
).reset_index()

# pivot df for population of people under 18
msa_tx_under18 = msa_tx.pivot(
    index=['msa_code','name'], 
    columns='year', 
    values='population_under_18'
).reset_index()

# create a df with just the migration to tyler 
msa_tx['total_in_migration'] = (
    msa_tx['moved_from_diff_county_same_state'] + 
    msa_tx['moved_from_diff_state'] + 
    msa_tx['moved_from_abroad']
)

msa_tx_migration = msa_tx.pivot(
    index=['msa_code','name'],
    columns='year',
    values=['lived_in_same_house', 
            'moved_within_county', 
            'moved_from_diff_county_same_state',
            'moved_from_diff_state',
            'moved_from_abroad',
            'total_in_migration']
) 

# flatten cols so theyre not stacked
msa_tx_migration.columns = [
    f"{var}_{year}" for var, year in msa_tx_migration.columns
]

# reset index
msa_tx_migration = msa_tx_migration.reset_index();
```

```python
print(len(msa_tx_pop))
print(len(msa_tx_age))
print(len(msa_tx_under18))
print(len(msa_tx_migration))
```

## Total population

```python
# create columns for population change in the past 5, 10, 15 and 20 years
msa_tx_pop['pct_change_19_24'] = ((msa_tx_pop[2024] - msa_tx_pop[2019]) / msa_tx_pop[2019]) * 100 # 5 years
msa_tx_pop['pct_change_14_24'] = ((msa_tx_pop[2024] - msa_tx_pop[2014]) / msa_tx_pop[2014]) * 100 # 10 years
msa_tx_pop['pct_change_09_24'] = ((msa_tx_pop[2024] - msa_tx_pop[2009]) / msa_tx_pop[2009]) * 100 # 15 years
msa_tx_pop['pct_change_05_24'] = ((msa_tx_pop[2024] - msa_tx_pop[2005]) / msa_tx_pop[2005]) * 100 # 19 years
```

```python
# past 19 years: Tyler metro area's population grew by 34% -- 13th out of 27 metro areas.
print("Population change 2005-24")
print(msa_tx_pop.sort_values(
        by='pct_change_05_24', 
        ascending=False, 
        na_position='last')
      .reset_index(drop=True)[['name','pct_change_05_24']])

# past 15 years: Tyler grew by 21.7% -- 14th out of 27 metro areas.
print("Population change 2009-24")
print(msa_tx_pop.sort_values(
        by='pct_change_09_24', 
        ascending=False, 
        na_position='last')
      .reset_index(drop=True)[['name','pct_change_09_24']])

# past 10 years: Tyler metro area's population grew by 13.8% -- 12th out of 29 metro areas
print("Population change 2014-24")
print(msa_tx_pop.sort_values(
        by='pct_change_14_24', 
        ascending=False, 
        na_position='last')
      .reset_index(drop=True)[['name','pct_change_14_24']])

# past 5 years: Tyler metro area's population grew by 7.02% -- 11th out of 29 metro areas.
print("Population change 2019-24")
print(msa_tx_pop.sort_values(
        by='pct_change_19_24', 
        ascending=False, 
        na_position='last')
      .reset_index(drop=True)[['name','pct_change_19_24']])
```

## Do the same for county-level

```python
# County-level data 
# ACS 1-year estimates are only available for counties with populations of 65,000 or more

counties_data = []

# loop through to get population data for all counties available in texas using ACS 1-year estimates.
# note that data is not availale for 2020.
for year in years:
    try:
        data = c.acs1.state_county(
            ('NAME', 
             'B01003_001E', # total population
             'B01002_001E', # median age
             'B09001_001E', # population under 18
             'B07003_004E', # Lived in the same house 1 year ago
             'B07003_007E', # Moved within same county
             'B07003_010E', # Moved from a different county in the same state
             'B07003_013E', # Moved from a different state
             'B07003_016E', # Moved from abroad 
            ),
            state_fips,
            Census.ALL,
            year=year
        )
        
        # Add year to each record
        for record in data:
            record['year'] = year
            counties_data.append(record)
        
        print(f"Retrieved data for {year}: {len(data)} counties")
    
    except Exception as e:
        print(f"Error for year {year}: {e}")

# make it a df
tx_counties = pd.DataFrame(counties_data)

# rename columns for readability
tx_counties = tx_counties.rename(columns={
    'NAME':'name',
    'B01003_001E': 'population',
    'B01002_001E': 'median_age',
    'B09001_001E': 'population_under_18',
    'B07003_004E': 'lived_in_same_house',
    'B07003_007E': 'moved_within_county',
    'B07003_010E': 'moved_from_diff_county_same_state',
    'B07003_013E': 'moved_from_diff_state',
    'B07003_016E': 'moved_from_abroad',
    'county':'county_code'
}
) 
```

```python
# create the same dfs as before, just with counties 
# pivot df so each column is a year
county_tx_pop = tx_counties.pivot(
    index=['county_code','name'], 
    columns='year', 
    values='population'
).reset_index()

# pivot df for median age 
county_tx_age = tx_counties.pivot(
    index=['county_code','name'], 
    columns='year', 
    values='median_age'
).reset_index()

# pivot df for population of people under 18
county_tx_under18 = tx_counties.pivot(
    index=['county_code','name'], 
    columns='year', 
    values='population_under_18'
).reset_index()

# create a df with just the migration to tyler 
tx_counties['total_in_migration'] = (
    tx_counties['moved_from_diff_county_same_state'] + 
    tx_counties['moved_from_diff_state'] + 
    tx_counties['moved_from_abroad']
)

tx_counties_migration = tx_counties.pivot(
    index=['county_code','name'],
    columns='year',
    values=['lived_in_same_house', 
            'moved_within_county', 
            'moved_from_diff_county_same_state',
            'moved_from_diff_state',
            'moved_from_abroad',
            'total_in_migration']
) 

# flatten cols so theyre not stacked
tx_counties_migration.columns = [
    f"{var}_{year}" for var, year in tx_counties_migration.columns
]

# reset index
tx_counties_migration = tx_counties_migration.reset_index();
```

```python
# create columns for population change in the past 5, 10, 15 and 20 years
county_tx_pop['pct_change_19_24'] = ((county_tx_pop[2024] - county_tx_pop[2019]) / county_tx_pop[2019]) * 100 # 5 years
county_tx_pop['pct_change_14_24'] = ((county_tx_pop[2024] - county_tx_pop[2014]) / county_tx_pop[2014]) * 100 # 10 years
county_tx_pop['pct_change_09_24'] = ((county_tx_pop[2024] - county_tx_pop[2009]) / county_tx_pop[2009]) * 100 # 15 years
county_tx_pop['pct_change_05_24'] = ((county_tx_pop[2024] - county_tx_pop[2005]) / county_tx_pop[2005]) * 100 # 19 years
```

```python
# past 19 years: Smith County's population also grew by 34% -- 30th out of 49 metro areas.
print("Population change 2005-24")
print(county_tx_pop.sort_values(
        by='pct_change_05_24', 
        ascending=False, 
        na_position='last')
      .reset_index(drop=True)[['name','pct_change_05_24']])

# past 15 years: Smith County grew by 21.7% -- 30th out of 50 metro areas.
print("Population change 2009-24")
print(county_tx_pop.sort_values(
        by='pct_change_09_24', 
        ascending=False, 
        na_position='last')
      .reset_index(drop=True)[['name','pct_change_09_24']])

# past 10 years: Smith County's population grew by 13.8% -- 29th out of 53 metro areas
print("Population change 2014-24")
print(county_tx_pop.sort_values(
        by='pct_change_14_24', 
        ascending=False, 
        na_position='last')
      .reset_index(drop=True)[['name','pct_change_14_24']])

# past 5 years: Tyler metro area's population grew by 7.02% -- 30th out of 54 metro areas.
print("Population change 2019-24")
print(county_tx_pop.sort_values(
        by='pct_change_19_24', 
        ascending=False, 
        na_position='last')
      .reset_index(drop=True)[['name','pct_change_19_24']])
```
